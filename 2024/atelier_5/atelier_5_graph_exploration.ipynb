### Énoncé du Problème
* Vous devez déterminez la valeur monétaire de ce qu'un fermier peut récolter dans un jeu de simulation.
* Le fermier ramasse tout les citrouilles qu'il peut atteindre sans traverser de bottes de foin.
* Les citrouilles valent:
    * S: 1\\$
    * M: 5\\$
    * L: 10\\$
    

### Les algorithmes d'explorations de graphes

#### Recherche en largeur


```plaintext
ParcoursLargeur(Graphe G, Sommet s):
    f ← CréerFile()

                f.enfiler(t)

#### Recherche en profondeur
C'est le même algorithme que la recherche en largeur, mais on utilise une pile au lieu d'une file.


### Implémentation

````python
# initialisation
    # lignes = int(input())
    # colonnes = int(input())
    # champ = []
    # for i in range(lignes):
    #     champ.append(input())
    # row_i = int(input())
    # col_i = int(input())
lignes = 6
colonnes = 6
champ = [
    "**LMLS",
    "S*LMMS",
    "S*SMSM",
    "******",
    "LLM*MS",
    "SSL*SS"
]
row_i, col_i = 5, 1

coord_i = (row_i, col_i)
to_visit, marked = set(), set()
to_visit.add(coord_i)
marked.add(coord_i)
````

````python
rep = 0
while to_visit:  # tant qu'il y a qqchose dans to_visit
    row,col = to_visit.pop()
    if champ[row][col] == "*":
        continue
    if champ[row][col] == "L":
        rep += 10
    elif champ[row][col] == "M":
        rep += 5
    elif champ[row][col] == "S":
        rep += 1

    for coord in {(row-1,col),(row+1,col), (row,col-1), (row,col+1)}:
        r,c = coord
        if r<0 or c<0 or c>=colonnes or r>=lignes:
            continue
        if coord not in marked:
            marked.add(coord)
            to_visit.add(coord)

print(rep)
````
