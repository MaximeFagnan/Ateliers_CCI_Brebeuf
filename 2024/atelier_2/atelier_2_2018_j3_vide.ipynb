### Transformer une entrée en liste de nombres entiers

Pour commencer, nous allons voir comment lire une entrée utilisateur composée de plusieurs nombres entiers séparés par des espaces, et la convertir en une liste de nombres entiers.

```python
# Exemple : lire une ligne d'entrée et la transformer en une liste de nombres entiers
entree = input()
liste_nombres = list(map(int, entree.split()))
print("Liste des nombres :", liste_nombres)
```

### Utiliser la fonction `sum()` et le slicing

La fonction `sum()` permet de calculer la somme des éléments d'une liste. Le slicing permet de sélectionner une sous-partie d'une liste.

```python
# Exemple : somme des éléments d'une liste et utilisation du slicing
liste = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Calculer la somme de toute la liste
somme_totale = sum(liste)
print("Somme de la liste :", somme_totale)

# Slicing : extraire les éléments de l'index 2 à 5 (exclus)
sous_liste = liste[:7]
print("Sous-liste (index 2 à 5 exclus) :", sous_liste)

# Somme des éléments de la sous-liste
somme_sous_liste = sum(sous_liste)
print("Somme de la sous-liste :", somme_sous_liste)
```

## Pretty Print

### Imprimer les éléments d'une liste sur une seule ligne

Pour imprimer les éléments d'une liste de nombres entiers sur une seule ligne, séparés par des espaces, nous pouvons utiliser la méthode `join()` sur une chaîne de caractères.

```python
# Exemple : imprimer les éléments d'une liste sur une seule ligne
liste_nombres = [10, 20, 30, 40, 50]
liste_chaine_charactere = map(str, liste_nombres)
chaine_charactere_avec_espace = " ".join(liste_chaine_charactere)
print(chaine_charactere_avec_espace)
```
